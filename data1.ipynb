## Understanding Peak Shifting and Its Importance

### What is Peak Shifting?
Peak shifting is a strategy used in energy management to redistribute energy consumption from peak demand periods to off-peak hours. It involves optimizing energy usage patterns to reduce strain on the power grid during times of high demand, typically during peak hours.

### Importance of Peak Shifting:
1. **Grid Reliability:** Peak shifting helps maintain grid stability and reliability by reducing the risk of blackouts or brownouts during peak demand periods.
   
2. **Cost Savings:** By shifting energy usage to off-peak hours, consumers and businesses can benefit from lower electricity rates, resulting in cost savings on energy bills.

3. **Environmental Impact:** Reducing peak energy demand can contribute to lowering greenhouse gas emissions and mitigating environmental impact associated with energy generation.

4. **Infrastructure Optimization:** Peak shifting helps optimize energy infrastructure utilization by spreading out energy demand more evenly throughout the day, reducing the need for costly infrastructure upgrades.

5. **Resource Conservation:** By reducing peak energy demand, peak shifting promotes efficient use of energy resources, contributing to overall resource conservation and sustainability.

6. **Energy Equity:** Implementing peak shifting strategies can enhance energy equity by ensuring fair and equitable access to affordable energy for all consumers, regardless of income level.

Peak shifting plays a crucial role in enhancing energy efficiency, promoting grid stability, and advancing sustainability goals, making it a valuable strategy for energy management in modern societies.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('output_2.csv')

# Group data by state
grouped_data = df.groupby('State')

# Plot pie chart for each state
for state, data in grouped_data:
    sizes = [data['peak_met'].sum(), data['peak_demand'].sum() - data['peak_met'].sum()]
    labels = ['Met', 'Not Met']

    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f'Peak Demand vs. Met - {state}')
    plt.axis('equal')
    plt.show()
```

## Peak Demand Analysis with Peak Shifting Model

### Introduction
This analysis visualizes peak demand vs. met demand for each state using a peak shifting model. The model redistributes unmet peak demand to off-peak hours, assuming a certain percentage can be shifted to meet the demand.

### Peak Shifting Model
The peak shifting model redistributes unmet peak demand to off-peak hours, assuming 80% of the unmet demand can be shifted to off-peak times.

### Implementation
1. **Loading Data**: Load peak demand data from the CSV file.
2. **Grouping by State**: Group data by state to analyze peak demand for each state separately.
3. **Peak Shifting Function**: Define a function `peak_shifting` to redistribute unmet demand to off-peak hours.
4. **Applying Peak Shifting**: Apply the `peak_shifting` function to each row of data for each state.
5. **Visualization**: Plot a pie chart for each state, showing the proportion of peak demand met with peak shifting.

### Peak Shifting Strategies
- **Demand Response Programs**: Incentivize consumers to shift energy usage to off-peak hours.
- **Time-of-Use Pricing**: Encourage energy usage during off-peak hours with lower prices.
- **Energy Storage Systems**: Store excess energy during off-peak hours for use during peak demand.
- **Smart Grid Technologies**: Monitor and manage energy consumption in real-time.
- **Public Awareness and Education**: Educate consumers about peak shifting benefits.

### Assumptions
- **Peak Shifting Percentage**: 80% of unmet peak demand is assumed to be shiftable to off-peak hours.

### Results Interpretation
The pie charts represent the proportion of peak demand met after applying the peak shifting model. Higher met demand indicates better peak load management.

### Actual Results (Example)
- **State**: Chandigarh
  - **Peak Demand**: 265 kW
  - **Peak Met**: 265 kW
  - **Peak Demand Met (With Peak Shifting)**: 212 kW
  - **Assumption**: 80% of unmet demand can be shifted to off-peak hours.

### Conclusion
The peak shifting model optimizes peak demand management, ensuring a higher proportion of peak demand is met by redistributing unmet demand. Effective implementation of peak shifting strategies and technologies can lead to a more resilient and sustainable energy system.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('output_2.csv')

# Group data by state
grouped_data = df.groupby('State')

# Define function for peak shifting
def peak_shifting(peak_demand, peak_met):
    if peak_demand > peak_met:
        # Calculate unmet peak demand
        unmet_demand = peak_demand - peak_met
        # Redistribute unmet demand to off-peak hours
        shifted_demand = peak_met + unmet_demand * 0.8  # Example: Shift 80% of unmet demand to off-peak hours
        return shifted_demand
    else:
        return peak_met

# Plot pie chart for each state with peak shifting
for state, data in grouped_data:
    data['shifted_peak'] = data.apply(lambda row: peak_shifting(row['peak_demand'], row['peak_met']), axis=1)
    sizes = [data['shifted_peak'].sum(), data['peak_demand'].sum() - data['shifted_peak'].sum()]
    labels = ['Met', 'Not Met']

    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f'Peak Demand vs. Met (with Peak Shifting) - {state}')
    plt.axis('equal')
    plt.show()
```
